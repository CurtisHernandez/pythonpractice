Some functions I made that I needed for these exercises (probably there is already a way to do this in numpy, but I defined it out myself in order to learn), plus some libraries:

```python
import numpy as np
import random
from IPython.display import Image

def vectorSum(listOfVecs): 
    newvec = []
    veclengths = [len(i) for i in listOfVecs]
    if veclengths[0] != int(sum(veclengths)/len(veclengths)):
        print("All the vectors must be of the same length")
    else:
        for j in range(len(listOfVecs[0])):
            newcoord = 0
            for k in range(len(listOfVecs)):
                newcoord += listOfVecs[k][j]
            newvec.append(newcoord)
    return newvec

def vectorScale(vector,scale):
    newvec = []
    for i in vector:
        newvec.append(i*scale)
    return newvec

# dot products are defined here as just having 2 vectors - any more and it's not a dot product, it's a tensor
def dotProduct(a,b):
    sum = 0
    for i in range(len(a)):
        sum += a[i] * b[i]
    return sum
```

Problem 1. Consider the vector v = (2, 1, 3, 7) and the vectors n1, n2, and n3 given by

n1 = (1, 2, 0, 0)

n2 = (0, 3, 1, 0)

n3 = (0, 7, 0, 1)  *pretend these are all pointy brackets, Jupyter doesn't like them

Suppose that w is orthogonal to v . Explain why w is a linear combination of {n1, n2, n3}.

For two vectors to be orthogonal, their dot product has to be equal to zero.  W would have to be some a, b, c, to scale n1, n2, and n3 respectively.  In other words:

w = a * (1, 2, 0, 0) + b * (0, 3, 1, 0) + c * (0, 7, 0, 1)
w = (a, 2a, 0, 0) + (0, 3b, b, 0) + (0, 7c, 0, c)
w = (a, 2a+3b+7c, b, c)

The dot product of v and w would be:

v1w1 + v2w2 + v3w3 + v4w4 = ?
2a + (-1)(2a + 3b + 7c) + 3b + 7c = ?
2a - 2a - 3b - 7c + 3b + 7c = 0

Since we don't even have to specify what a, b, and c would be to find that the dot product is zero, we can assume that the w which is orthogonal to v will be a linear combination of {n1, n2, n3}.

```python
a = random.randint(1,10)
print("The weight a in the linear combination = " + str(a))
b = random.randint(1,10)
print("The weight b in the linear combination = " + str(b))
c = random.randint(1,10)
print("The weight c in the linear combination = " + str(c))

v = [2, -1, 3, 7]
n1 = [1, 2, 0, 0]
n2 = [0, 3, 1, 0]
n3 = [0, 7, 0, 1]

print("The linear combination of n1, n2, and n3 with weights a, b, and c is:")
print(np.column_stack([vectorSum([vectorScale(n1,a),vectorScale(n2,b),vectorScale(n3,c)])]))
print("And lo and behold, the dot product of that and vector <2, -1, 3, 7> is:")
print(str(dotProduct(v,vectorSum([vectorScale(n1,a),vectorScale(n2,b),vectorScale(n3,c)]))))
```
